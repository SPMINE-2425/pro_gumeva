```python
import pandas as pd
import requests
import json
 
# Ejemplo de solicitud a la API
response = requests.get('https://www.datos.gov.co/resource/72sg-cybi.json')
base_deptos = response.json()
 
# Muestra los primeros elementos del JSON para verificar su estructura
print(json.dumps(base_deptos, indent=4))

if base_deptos:
    df = pd.DataFrame(base_deptos)
    print("Primeras filas del DataFrame:")
    display(df.head())
else:

    print("No se recibieron datos.")
```
